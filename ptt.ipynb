```python
import requests
response = requests.get("https://www.ptt.cc/bbs/movie/M.1539183746.A.DF0.html")
html = BeautifulSoup(response.text)
# html
```

```python
main = html.find("div", id="main-content")
# 把等等要丟的東西的內容先保留起來
val = main.find_all("span", class_="article-meta-value")
print("作者:", val[0].text)
print("標題:", val[2].text)
print("時間:", val[3].text)
# 開始丟東西
metas = main.find_all("div", class_="article-metaline")
for m in metas:
    m.extract()
metas = main.find_all("div", class_="article-metaline-right")
for m in metas:
    m.extract()
# 處理推噓文
push = main.find_all("div", class_="push")
score = 0
for p in push:
    pcontent = p.find("span", class_="push-tag").text
    if "推" in pcontent:
        score = score + 1
    elif "噓" in pcontent:
        score = score - 1
    # 算完score, 順勢把它丟掉
    p.extract()
print("分數:", score)
# 阻礙已被清除,可以拿裡面的文字了
print("內容:", main.text)
```
