```python
from urllib.request import urlopen
import json

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

response = urlopen("https://www.google.com/doodles/json/2018/9?hl=zh_TW")
doodles = json.load(response)
doodles
```

```python
# doodles -> List d->Dict d["url"]->String
for d in doodles:
    url = "https:" + d["url"]
    print(d["title"], url)
    # open圖網址
    response = urlopen(url)
    img = response.read()
    # 存檔
    path = "doodles/" + url.split("/")[-1]
    f = open(path, "wb")
    f.write(img)
    f.close()
```

```python
from urllib.request import urlopen, urlretrieve
import json

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

response = urlopen("https://www.google.com/doodles/json/2018/10?hl=zh_TW")
doodles = json.load(response)
# doodles -> List d->Dict d["url"]->String
for d in doodles:
    url = "https:" + d["url"]
    print(d["title"], url)
    path = "doodles/" + url.split("/")[-1]

    urlretrieve(url, path)
```

整年


```python
from urllib.request import urlopen, urlretrieve
import json
import os

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

for m in range(1, 13):
    monthurl = "https://www.google.com/doodles/json/2018/" + str(m) + "?hl=zh_TW"
    print(monthurl)
    response = urlopen(monthurl)
    doodles = json.load(response)
    # 創資料夾(mkdir前要檢查)
    monthdir = "doodles/" + str(m)
    if not os.path.exists(monthdir):
        print("幫你創", monthdir)
        os.mkdir(monthdir)
        
    for d in doodles:
        url = "https:" + d["url"]
        print(d["title"], url)
        path = monthdir + "/" + url.split("/")[-1]

        urlretrieve(url, path)
```
