```python
from urllib.request import urlopen
import json

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

response = urlopen("https://www.google.com/doodles/json/2018/9?hl=zh_TW")
doodles = json.load(response)
doodles
```

```python
# doodles -> List d->Dict d["url"]->String
for d in doodles:
    url = "https:" + d["url"]
    print(d["title"], url)
    # open圖網址
    response = urlopen(url)
    img = response.read()
    # 存檔
    path = "doodles/" + url.split("/")[-1]
    f = open(path, "wb")
    f.write(img)
    f.close()
```

組合

```python
from urllib.request import urlopen, urlretrieve
import json

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

response = urlopen("https://www.google.com/doodles/json/2018/10?hl=zh_TW")
doodles = json.load(response)
# doodles -> List d->Dict d["url"]->String
for d in doodles:
    url = "https:" + d["url"]
    print(d["title"], url)
    path = "doodles/" + url.split("/")[-1]

    urlretrieve(url, path)
```
